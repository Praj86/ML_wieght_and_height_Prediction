```python
#importing the libararies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset= pd.read_csv("height.csv")
```

```python
dataset.head()
```

```python
#3. checking for outliers
fig, axs = plt.subplots(3, figsize= (5,5))
plt1 = sns.boxplot(dataset['age'], ax = axs[0])
plt2 = sns.boxplot(dataset['height'], ax = axs[1])
plt.tight_layout()
                   
```

```python
#exploratory Data Analysis
#1. Distribution of the target variable
sns.displot(dataset['height']);
```

```python
#2. How height are related with other variables
sns.pairplot(dataset, x_vars=['age'], y_vars='height',height=4, aspect=1, kind='scatter')
plt.show()
```

```python
#3. heatmap
sns.heatmap(dataset.corr(), annot=True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
#setting the value for x and y
x= dataset[['age']]
y= dataset['height']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state = 50)
```

```python
slr= LinearRegression()
slr.fit(x_train,y_train)
```

```python
#printing the model coefficients
print('Intercept: ', slr.intercept_)
print('coefficient:', slr.coef_)
```

```python
print('Regression Equation: Sales= 32.83 + 0.5 * AGE')
```

```python
#line of best fit
plt.scatter(x_train,y_train)
plt.plot(x_train, 32.83 + 0.59 * x_train, 'r')
plt.show()
```

```python
#Prediction of Test and Training set result
y_pred_slr=slr.predict(x_test)
x_pred_slr=slr.predict(x_test)
```

```python
print("Prediction for test set: {} ". format(y_pred_slr))
```

```python
#Actual value and the predict value
slr_diff = pd.DataFrame({'Actual Value': y_test, 'Predicted Value': y_pred_slr})
slr_diff
```
